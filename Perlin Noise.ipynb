```python
import matplotlib.pyplot as plt
import numpy as np

def linear_congruential_generator(seed, mul, mod):
    return (seed * mul) % mod

def generate(x, _seed):
    mod = 4294967291
    mul = 1588635695
    seed = linear_congruential_generator(_seed, mul, mod)
    nums = np.array([])
    for i in range(x):
        seed = linear_congruential_generator(seed, mul, mod)
        nums = np.append(nums, seed / mod)
    return nums

def plot(seed, x):
    nums = generate(x*x, seed)
    arr = np.reshape(nums, (x,x))
    plt.imshow(arr, origin='upper', vmin=0, vmax=1)

plot(1, 50)
```
